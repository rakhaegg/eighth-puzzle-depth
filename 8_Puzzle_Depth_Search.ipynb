**Setiap Tile memiliki state atau node**

```python
class Tile:
  def __init__(self , state=None , location=None):
    self.left = None
    self.right = None
    self.up = None
    self.bottom = None
    self.state = state
    self.location = location
```

**Keadaan puzzle akan direpresentasikan dengan Objek sendiri**

```python
import numpy as np
```

```python
class Puzzle:
  def __init__(self , panjang , lebar):
    self.phi = None
    # Array data menyimpan Objek Tile
    self.data = []
    self.panjang = panjang
    self.lebar = lebar
  def getNodeContainsZeroState(self):
    for x in range(self.data.shape[0]):
      for y in range(self.data.shape[1]):
        if self.data[x][y].state == 0:
          return self.data[x][y]
  def changeTile(self , tileFirst , tileSecond):
    tempFirst = None
    tempSecond = None
    for x in range(self.data.shape[0]):
      for y in range(self.data.shape[1]):
        if self.data[x][y].state == tileFirst:
          tempFirst = self.data[x][y]
        if self.data[x][y].state == tileSecond:
          tempSecond = self.data[x][y]
    
    temp = tempFirst.state
    tempFirst.state = tempSecond.state
    tempSecond.state = temp
  def getInformation(self):
    tempDict = {}
    for x in range(self.data.shape[0]):
      for y in range(self.data.shape[1]):
        tempDict[self.data[x][y].state] = self.data[x][y].location
    return tempDict
  def setTile(self , data):
    for x in data.keys():
      state = Tile(x , data[x])
      self.data.append(state)
    arr = np.array(self.data)
    result = np.reshape(arr , (self.lebar , self.panjang))
    self.data= result
    # Konfigurasi Location
    for x in range(self.data.shape[0]):
      for y in range(self.data.shape[1]):
        if x == 0 and y == 0:
          # Position 1 
          self.data[x][y].left = None
          self.data[x][y].up = None
          self.data[x][y].bottom = self.data[x+1][y]
          self.data[x][y].right = self.data[x][y+1]
        elif x == 0 and y == (self.panjang-1):
          # Positino 3 
          self.data[x][y].up = None
          self.data[x][y].bottom = self.data[x+1][y]
          self.data[x][y].left = self.data[x][y-1]
          self.data[x][y].right = None
        elif x == 0 and y != 0 and y != (self.panjang - 1):
          # Position Between 3 and 1
          self.data[x][y].up = None
          self.data[x][y].bottom = self.data[x+1][y]
          self.data[x][y].left = self.data[x][y-1]
          self.data[x][y].right = self.data[x][y+1]
        elif x != 0 and x != (self.lebar -1 ) and y == 0:
          # Position 4
          self.data[x][y].up = self.data[x-1][y]
          self.data[x][y].bottom = self.data[x+1][y]
          self.data[x][y].left = None
          self.data[x][y].right = self.data[x][y+1]
        elif x != 0 and x != (self.lebar - 1) and y!= 0 and y!= (self.panjang-1):
          # Position Mid 4 and 6
          self.data[x][y].up = self.data[x-1][y]
          self.data[x][y].bottom = self.data[x+1][y]
          self.data[x][y].left = self.data[x][y-1]
          self.data[x][y].right = self.data[x][y+1]
        elif x != 0 and x != (self.lebar - 1) and y == (self.panjang -1 ):
          # Positon 6
          self.data[x][y].up = self.data[x-1][y]
          self.data[x][y].bottom = self.data[x+1][y]
          self.data[x][y].left = self.data[x][y-1]
          self.data[x][y].right = None
        elif x == (self.lebar - 1) and y == 0:
          # Position 7
          self.data[x][y].up  = self.data[x-1][y]
          self.data[x][y].bottom = None
          self.data[x][y].left = None
          self.data[x][y].right = self.data[x][y+1]
        elif x == (self.lebar - 1) and y!=0 and y!= (self.panjang -1):
          # Position Between 7 and 9
          self.data[x][y].up = self.data[x-1][y]
          self.data[x][y].bottom = None
          self.data[x][y].left = self.data[x][y-1]
          self.data[x][y].right = self.data[x][y+1]
        elif x == (self.lebar - 1) and y == (self.panjang - 1):
          # Position 9
          self.data[x][y].up = self.data[x-1][y]
          self.data[x][y].bottom = None
          self.data[x][y].left = self.data[x][-1]
          self.data[x][y].right = None
        

  def check(self):
    for x in range(self.data.shape[0]):
      for y in range(self.data.shape[1]):
        print("Location : " , self.data[x][y].location )
        node = self.data[x][y]
        if node.up:
          print("UP : " , self.data[x][y].up.location )
        if node.bottom:
          print("BOTTOM : " ,self.data[x][y].bottom.location )
        if node.left:  
          print("LEFT : " , self.data[x][y].left.location)
        if node.right:
          print("RIGHT : " , self.data[x][y].right.location)
      print()
```

```python
def checkIfExists(explored , current):
  for x in explored:
    if x.getInformation() == current.getInformation():
      return False
  return True
def do_depth_search(start_state ,  goal):
  p = Puzzle(3 , 3)
  p.setTile(start_state)

  q = LifoQueue()
  
  q.put(p)
  explored = []
  explored.append(p)
  count = 0
  while q.qsize() != 0:
    
    parent = q.get()

    nodeParent = parent.getNodeContainsZeroState()

    tempxx = parent.getInformation()
    if tempxx[0] == goal[0] and tempxx[1] == goal[1] and tempxx[2] == goal[2]:
      last = explored[-1]
      while last : 
        print(last.getInformation())
        last = last.phi 
      break
    if nodeParent.up:
      childUp = Puzzle(3 , 3)
      childUp.setTile(parent.getInformation())      
      #childUp = copy.deepcopy(parent)
      childUp.changeTile(0 , nodeParent.up.state)
      if checkIfExists(explored , childUp):
        childUp.phi = parent
        q.put(childUp)
        explored.append(childUp)
    if nodeParent.bottom:
      childBottom = Puzzle(3 , 3)
      childBottom.setTile(parent.getInformation())
      #childBottom = copy.deepcopy(parent)
      childBottom.changeTile(0 , nodeParent.bottom.state)
      if checkIfExists(explored , childBottom):
        childBottom.phi = parent
        q.put(childBottom)
        explored.append(childBottom)
    if nodeParent.left:
      childLeft = Puzzle(3 , 3)
      childLeft.setTile(parent.getInformation())
      #childLeft = copy.deepcopy(parent)
      childLeft.changeTile(0 , nodeParent.left.state)
      if checkIfExists(explored , childLeft):
        childLeft.phi = parent
        q.put(childLeft)
        explored.append(childLeft)
      #print(nodeParent.left.state)
    if nodeParent.right:
      childRight = Puzzle(3 , 3)
      childRight.setTile(parent.getInformation())
      #childRight = copy.deepcopy(parent)
      childRight.changeTile(0 , nodeParent.right.state)
      if checkIfExists(explored , childRight):
        childRight.phi = parent
        q.put(childRight)
        explored.append(childRight)
      #print(nodeParent.right.state)
    count = count+1

do_depth_search(start_state , goal_state)
```
